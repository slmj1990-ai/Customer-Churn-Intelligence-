```python
import pandas as pd
from sqlalchemy import create_engine, text

# conexión
engine = create_engine("mysql+pymysql://root:Leticia.1@localhost/ecommerce_db")
```

```python
#Cargando datos desde MySQL
df = pd.read_sql("SELECT * FROM ecommerce_customers", engine)

df.head()
```

```python
def save_table(df_table, table_name):

    with engine.connect() as conn:
        conn.execute(text(f"DROP TABLE IF EXISTS {table_name}"))

    df_table.to_sql(
        table_name,
        con=engine,
        index=False
    )

    print(f"Tabla creada: {table_name}")
```

Dashboard 2 - Clientes por país

```python
customers_country = (
    df.groupby("country")
    .size()
    .reset_index(name="customers")
    .sort_values("customers", ascending=False)
)

save_table(customers_country, "dashboard_customers_country")
```

Dashboard 3 - Clientes por género 

```python
customers_gender = (
    df.groupby("gender")
    .size()
    .reset_index(name="customers")
)

save_table(customers_gender, "dashboard_customers_gender")
```

```python
customers_gender = (
    df.groupby("gender")
    .size()
    .reset_index(name="customers")
)

save_table(customers_gender, "dashboard_customers_gender")
```
